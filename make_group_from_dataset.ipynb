```
PATH = "/content/drive/My Drive/IE481 데이터 가시화/dataset.csv"
df = pd.read_csv(PATH)
df.head()
```

```
df = df.sort_values(by=['time'])
df
```

```
pd.Timestamp(df.loc[0, 'time'])
```

```
df["new_time"] = df.apply(lambda row:pd.Timestamp(row.time), axis=1)
df["time_sort"] = df.apply(lambda row:row.time.split()[-1], axis=1)
df["day_of_week"] = df.apply(lambda row:row.new_time.dayofweek, axis=1)
df["new_time_group"] = df.apply(lambda row:row.new_time.strftime('%r'), axis=1)
df
```

```
df['Exercise'] = df['Exercise'].fillna(0)
df['Exercise'] = df['Exercise'].astype(float)
df['Exercise'] = df['Exercise'].astype(int)
df['Not Classified'] = df['Not Classified'].fillna(0)
df['Not Classified'] = df['Not Classified'].astype(int)
```

```
newdf = df.groupby(['day_of_week', 'time_sort', 'new_time_group']).sum()
newdf
```
