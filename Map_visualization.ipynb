```
dataset_directory = '/content/drive/My Drive/IE481 데이터 가시화/'

drive.mount('/content/drive')

location = pd.read_csv(join(dataset_directory, 'dataset.csv'))
location.head()
```

```
location_sort = location.sort_values('time')
location_index = location_sort
location_index['index_col'] = location_index['userId'] + " "+ location_index['time']
location_index.set_index('index_col', drop= True, inplace = True)
activities = location_index.filter(items=["Exercise", "SNS", "Video", "Sleep", "Not Classified"])
location_index.head()
```

```
location_index
```

```
location_index['activity'] = activities.idxmax(axis=1)
location_index
```

```
import plotly.express as px
import plotly

import matplotlib.pyplot as plt

fig_map = px.scatter_mapbox(location_index, lat="latitude", lon="longitude", hover_name = 'activity',
                        animation_frame='time', center=dict({'lat': 36.37, 'lon': 127.36}), zoom=13, color = 'activity'
            )

fig_map.update_layout(mapbox_style="carto-positron",width=900,
    height=700)
fig_map.layout.sliders[0].currentvalue.xanchor="left"
fig_map.layout.sliders[0].currentvalue.offset=-100
fig_map.layout.sliders[0].currentvalue.prefix=""
fig_map.layout.sliders[0].len=.9
fig_map.layout.sliders[0].currentvalue.font.color="indianred"
fig_map.layout.sliders[0].currentvalue.font.size=20
fig_map.layout.sliders[0].y= 1.1
fig_map.layout.sliders[0].x= 0.15
fig_map.show()
plotly.offline.plot(fig_map, filename='/content/drive/My Drive/abcd.html')
```
